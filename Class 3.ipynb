```python
# Simplified implementation of Queue (using built-ins)
class Queue:
    def __init__(self):
        self.items = []
    def enqueue(self,value):
        self.items.insert(0,value)
    def dequeue(self):
        return self.items.pop()

    #nice to have methods
    def size(self):
        return len(self.items)
    def peek(self):
        return self.items[len(self.items)-1]
    def is_empty(self):
        return self.items == []
```

```python
class QueueII:
    class __Node:
        def __init__(self,data):
            self.data=data
            self.next=None
            #Create a new node with the provided data
            #Set the next pointer to None
    def __init__(self):
        self.back=None
        self.front=None
        self._size=0
        #Initialize front and back pointers to None
        #Initialize size (_size) to 0
    def enqueue(self,data):
        new_node = self.__Node(data)
        if self.front is None:
            self.front = new_node
            self.back = new_node
        else:
            self.back.next = new_node
            self.back = new_node
        self._size +=1
        #First create a new node
        # If the queue is empty front and back is None
        # If not Update the next pointer of the back node to the new one and update back a new_node
        # Finally increment in 1 the size
    def print_queue(self):
        tem=self.front
        while tem is not None:
            print(tem.data)
            tem = tem.next
        return
        #Initializing a tem pointer in front
        #As long as tem is not None:
        ##Print the data of the current node (tem)
        #Move tem to the next node
        
    def dequeue(self):
        tem = self.front
        if self.front is None:
            raise IndexError("The queue is empty")
            # if front is None return None ( the Queue is empty)
        if self.front == self.back:
            self.front = None
            self.back = None
            # if front is equal to back, front and back is None
        else:
            self.front = tem.next
            tem.next = None
            
        self._size-=1
        return tem.data
            #Decrement size by 1
            #Return data from tem node

    def size(self):
        return self._size
         #Return current queue size (_size)
        
    def peek(self):
        if self.front is None:
            return None
        #if front is None return None (the queue is empty)
        else:
        #if not return the data from node front
            return self.front.data
    def is_empty(self):
        return self.front is None
        #Return True if front is None, False otherwise

myQueue = QueueII()
print('Empty:', myQueue.is_empty())
myQueue.enqueue(1)
myQueue.enqueue(2)
myQueue.enqueue(3)
myQueue.print_queue()
print('Size:', myQueue.size())
print('Empty:', myQueue.is_empty())
print('Peek:', myQueue.peek())
myQueue.dequeue()
myQueue.print_queue()

print('back:',myQueue.back.data)
```

```python
class SinglyLindekList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
        #Challenge for you: see if you can have a tail ref and how that impacts performance

    def append(self,value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next=new_node
            self.tail=new_node
        
        self.length+=1
            
    def insert(self,i,value):
        #This method has to insert a new node (with value) Before the index, not after
        new_node = self.__Node(value)
        if i<=1:
            new_node.next = self.head
            self.head = new_node
            self.length +=1
            return new_node
        elif i>self.length:
            self.append(value)
        else:
            tem = self.head
            for node in range(1, i-1):
                tem = tem.next
            new_node.next = tem.next
            tem.next = new_node
            self.length +=1
        
            

    def __str__(self):
        #this dunder method must return a string at all times

        out = '['
        current = self.head
        if current:
            out += '%s' % repr(current.data)
            current = current.next
            while current:
                out += ', %s' % repr(current.data)
                current = current.next
        out +=']'
        return out


ll = SinglyLindekList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.insert(0,100)
ll.insert(100,500)
ll.insert(5,200)
ll.insert(2,'deiby')
ll.insert(7,'sisas')
print(ll)
```
